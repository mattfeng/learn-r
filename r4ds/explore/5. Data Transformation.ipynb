## Filter

- Use `near()` for floating point comparisons.
- Use `%in%` for comparisons involving many "or" operations
- Don't use `&&` or `||`; those operations will perform conditional execution, which will not give the desired bitmasks we need.
- Use `is.na(x)` to test for NA.

```r
jan1 = filter(flights, month == 1, day == 1)
print(jan1)
```
